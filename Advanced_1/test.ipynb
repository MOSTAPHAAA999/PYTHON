```python
import pandas as pd 
import matplotlib.pyplot as plt 
from datasets import load_dataset
#load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States']
df_us
```

```python
df_us['job_posted_month']=df_us['job_posted_date'].dt.strftime('%B')
df_us
```

```python
df_us_pivot =df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_us_pivot
```

```python
df_us_pivot.reset_index(inplace=True)
df_us_pivot['job_posted_month_num'] =pd.to_datetime(df_us_pivot['job_posted_month'], format ='%B').dt.month
df_us_pivot.sort_values('job_posted_month_num', inplace=True)  
df_us_pivot.set_index('job_posted_month', inplace=True)
df_us_pivot.drop(columns='job_posted_month_num', inplace=True)
df_us_pivot
```

```python
top_5_jobs=df_us_pivot.sum().sort_values(ascending=False).head(5).index
top_5_jobs.tolist()
```

```python
df_ss =df_us_pivot.sum()
df_ss.sort_values(ascending=False, inplace=True)
df_ss
```

```python
df_us_pivot[top_5_jobs].plot(kind= 'bar', figsize=(15,10))
```

```python
df_India= df[df['job_country']== 'India'].copy()
df_India
```

```python
df_India['job_posted_month']= df_India['job_posted_date'].dt.strftime('%B')
df_India
```

```python
df_India_pivot = df_India.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc= 'size')
df_India_pivot
```

```python
df_India_pivot.reset_index(inplace= True)
df_India_pivot['job_posted_month_num']= pd.to_datetime(df_India_pivot['job_posted_month'], format='%B').dt.month
df_India_pivot.sort_values('job_posted_month_num', inplace=True)
df_India_pivot.set_index('job_posted_month' , inplace=True)
df_India_pivot.drop(columns='job_posted_month_num')
df_India_pivot
```

```python
df_AI = df_India_pivot.merge(df_us_pivot,on='job_posted_month', suffixes=('United States', 'India'))
df_AI
```

```python
top_5_jobs=df_AI.sum().sort_values(ascending=False).head(5)
top_5_jobs.to_list
```
