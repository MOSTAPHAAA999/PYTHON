```python
import pandas as pd
import matplotlib.pyplot as plt 
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_Sudan = df[df['job_country'] == 'Sudan']
df_Sudan
```

```python
df_Sudan['job_posted_month'] = df_Sudan['job_posted_date'].dt.strftime('%B')
df_Sudan
```

```python
df_Sudan_pivot  = df_Sudan.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_Sudan_pivot
```

```python
df_Sudan_pivot.reset_index(inplace=True)
df_Sudan_pivot['job_posted_NUM']=pd.to_datetime(df_Sudan_pivot['job_posted_month'], format='%B').dt.month
df_Sudan_pivot.sort_values('job_posted_NUM', inplace=True)
df_Sudan_pivot.set_index('job_posted_month', inplace=True)
df_Sudan_pivot.drop(columns='job_posted_NUM', inplace=True)
df_Sudan_pivot
```

```python
Top_5 = df_Sudan_pivot.sum().sort_values(ascending=False).head(5).index
df_Sudan_pivot[Top_5].plot(kind='bar', figsize=(10, 5))
plt.show()
```

```python
df_Mexico = df[df['job_country' ] == 'Mexico']
df_Mexico
```

```python
df_Mexico['job_posted_month'] = df_Mexico['job_posted_date'].dt.strftime('%B')
df_Mexico
df_Mexico_pivot =df_Mexico.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_Mexico_pivot
```

```python
df_Mexico_pivot.reset_index(inplace=True)
df_Mexico_pivot['job_posted_NUM']=pd.to_datetime(df_Mexico_pivot['job_posted_month'], format='%B').dt.month
df_Mexico_pivot.sort_values('job_posted_NUM', inplace=True) 
df_Mexico_pivot.set_index('job_posted_month', inplace=True)
df_Mexico_pivot.drop(columns='job_posted_NUM', inplace=True)
df_Mexico_pivot
```

```python
df_con = pd.concat([df_Mexico_pivot, df_Sudan_pivot])
df_con
```

```python
df_merged =pd.merge(df_Mexico_pivot, df_Sudan_pivot, on='job_posted_month', suffixes=('_Mexico', '_Sudan'))
df_merged
```

```python
top_5_jobs=df_merged.sum().sort_values(ascending=False).head(5)
top_5_jobs.to_list
```

```python
df_merged[top_5_jobs.index].plot(kind='bar', figsize=(20, 15))
plt.show()
```
