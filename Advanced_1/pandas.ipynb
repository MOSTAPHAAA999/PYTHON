```python
import pandas as pd 
import matplotlib.pyplot as plt 
from datasets import load_dataset
#load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_Germany = df[df['job_country'] == 'Germany'].copy()
df_Germany
```

```python
df_mexico = df[df['job_country'] == 'Mexico'].copy()
df_mexico
```

```python
df_Germany['job_posted_month']=df_Germany['job_posted_date'].dt.strftime('%B')
df_Germany
```

```python
df_Germany_pivot = df_Germany.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_Germany_pivot
df_mexico_pivot = df_mexico.pivot_table(index ='job_posted_month', columns='job_title_short',aggfunc='size')
df_mexico_pivot
```

```python
df_Germany_pivot.reset_index(inplace=True)
df_Germany_pivot['job_posted_month_no'] = pd.to_datetime(df_Germany_pivot['job_posted_month'], format='%B').dt.month
df_Germany_pivot.sort_values('job_posted_month_no', inplace=True)
df_Germany_pivot.set_index('job_posted_month', inplace=True)
df_Germany_pivot.drop(columns='job_posted_month_no', inplace=True)
df_Germany_pivot
```

```python
df_mexico_pivot.reset_index(inplace=True)
df_mexico_pivot['job_posted_month_numb'] = pd.to_datetime(df_mexico_pivot['job_posted_month'], format ='%B').dt.month
df_mexico_pivot.sort_values('job_posted_month_numb', inplace=True)
df_mexico_pivot.set_index('job_posted_month', inplace=True)
df_mexico_pivot.drop(columns='job_posted_month_numb', inplace=True)
df_mexico_pivot
```

```python
df_MG = df_Germany_pivot.merge(df_mexico_pivot, on='job_posted_month', suffixes=('_Germany', '_Mexico'))
df_MG
```

```python
Top_5_JOBS = df_MG.sum().sort_values(ascending=False).head(5).index
Top_5_JOBS.tolist()
```

```python
df_MG[Top_5_JOBS].plot(kind='bar', figsize=(15,10))
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
months = df['job_posted_month'].unique()
months 
```

```python
dict_months = {months : df[df['job_posted_month'] == months] for months in months}
dict_months
```

```python
dict_months.keys()
```

```python
Q1 = [dict_months['Jan'], dict_months['Feb'], dict_months['Mar']]
df_Q1 = pd.concat(Q1,ignore_index=True)
df_Q1
```

```python
Q2 = [dict_months['Apr'], dict_months['May'], dict_months['Jun']]
df_Q2 = pd.concat(Q2,ignore_index=True)
df_Q2
```

```python
Q1_Q2 = pd.concat([df_Q1, df_Q2], ignore_index=True)
Q1_Q2
```

```python
Q1_Q2['job_posted_month'].value_counts().plot(kind='barh' , color = 'blue', figsize=(10, 5) , title = 'Number of Jobs Posted in Q1')
plt.show()
```

```python
df_Q1_pivot = df_Q1.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_Q1_pivot.reset_index(inplace=True)
df_Q1_pivot['job_posted_month_no'] = pd.to_datetime(df_Q1_pivot['job_posted_month'], format='%b').dt.month
df_Q1_pivot.sort_values('job_posted_month_no', inplace=True)
df_Q1_pivot.set_index('job_posted_month', inplace=True)
df_Q1_pivot.drop(columns='job_posted_month_no', inplace=True)
df_Q1_pivot
```

```python
df_Q2_pivot =df_Q2.pivot_table(index = 'job_posted_month',columns='job_title_short',aggfunc='size')
df_Q2_pivot.reset_index(inplace= True)
df_Q2_pivot['job_posted_month_no'] = pd.to_datetime(df_Q2_pivot['job_posted_month'], format='%b').dt.month
df_Q2_pivot.sort_values('job_posted_month_no', inplace=True)
df_Q2_pivot.set_index('job_posted_month', inplace=True)
df_Q2_pivot.drop(columns='job_posted_month_no', inplace=True)
df_Q2_pivot

```

```python
df_Q1_pivot['job_posted_month'] = df_Q1_pivot.index
df_Q2_pivot['job_posted_month'] = df_Q2_pivot.index
df_Q1_pivot.drop(columns='job_posted_month', inplace=True)
df_Q2_pivot.drop(columns='job_posted_month', inplace=True)
df_Q1_pivot
```

```python
Q_S = df_Q1_pivot.merge(df_Q2_pivot, on='job_posted_month',how='outer', suffixes=('_Q1', '_Q2'))
Q_S.reset_index(inplace= True)
Q_S['job_posted_month_no'] = pd.to_datetime(Q_S['job_posted_month'], format='%b').dt.month
Q_S.sort_values('job_posted_month_no', inplace=True)
Q_S.set_index('job_posted_month', inplace=True)
Q_S.drop(columns='job_posted_month_no', inplace=True)
Q_S.fillna(0, inplace=True)
Q_S
```

```python
# Plot the bar chart
Q_S.plot(kind='bar', figsize=(15, 10), title='Number of Jobs Posted in Q1 and Q2')
# Customize the plot
plt.title('Number of Jobs Posted in Q1 and Q2')  # Set the title
plt.legend(loc='upper center', bbox_to_anchor=(.5, 1), ncol=4)  # Position the legend above the plot, horizontal
plt.ylim(0, 30000)                              # Set the y-axis limit
# Show the plot
plt.show()
```
