```python
import pandas as pd 
import matplotlib.pyplot as plt
from datasets import load_dataset

#load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_Germany = df[df['job_country'] == 'Germany'].copy()
```

```python
print(df_Germany['job_posted_date'].dtype)
```

```python
df_Germany['job_posted_month']=df_Germany['job_posted_date'].dt.strftime('%B')
df_Germany
```

```python
df_Germany_pivot = df_Germany.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_Germany_pivot
```

```python
df_Germany_pivot.reset_index(inplace=True)
df_Germany_pivot['job_posted_month_no'] = pd.to_datetime(df_Germany_pivot['job_posted_month'], format='%B').dt.month
df_Germany_pivot.sort_values('job_posted_month_no', inplace=True)
df_Germany_pivot.set_index('job_posted_month', inplace=True)
df_Germany_pivot.drop(columns='job_posted_month_no', inplace=True)
df_Germany_pivot
```

```python
df_Germany_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')

df_Germany_software_pivot
```

```python
df_merged = df_Germany_pivot.merge(df_Germany_software_pivot, on='job_posted_month')
df_merged
```

```python
Top_5_Jobs = df_merged.sum().sort_values(ascending=False).head(5).index
Top_5_Jobs.tolist()
```

```python
df_merged[Top_5_Jobs].plot(kind='barh', figsize=(20, 8),color=['#522258', '#4158A6', '#C63C51', '#D95F59', '#4158A6'])
plt.title('Top 5 Jobs in Germany')
plt.xlabel('Month')
plt.ylabel('Number of Jobs')
plt.show()
```

```python
df_merged[Top_5_Jobs].plot(kind='line', figsize=(20, 8),color=['#522258', '#4158A6', '#C63C51', '#D95F59', '#4158A6'])
plt.ylabel('Number of Jobs')
plt.ylim(0, 20000)
plt.xlabel(2023)
plt.title('Top 5 Jobs in Germany') 
plt.legend(loc='upper right')
plt.show()
```
