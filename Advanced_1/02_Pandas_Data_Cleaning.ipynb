```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.value_counts
```

```python
df.loc[10:,'salary_hour_avg':'salary_year_avg']
```

```python
Hourly_salary_median = df['salary_hour_avg'].median()
```

```python
Annual_salary_median=df['salary_year_avg'].median()
```

```python
df_filled=df
```

```python
df_filled['salary_year_avg'].fillna(Annual_salary_median, inplace=True)
df_filled['salary_hour_avg'].fillna(Hourly_salary_median, inplace=True)
```

```python
df_unique = df_filled.drop_duplicates(subset=['job_title','company_name'])
df_unique.info()
```
