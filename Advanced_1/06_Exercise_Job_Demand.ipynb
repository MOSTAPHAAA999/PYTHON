```python
import pandas as pd 
import matplotlib.pyplot as plt
from datasets import load_dataset

#load the dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_Germany = df[df['job_country'] == 'Germany'].copy()
df_Germany
```

```python
print(df_Germany['job_posted_date'].dtype)
df_Germany['job_posted_date'] = pd.to_datetime(df_Germany['job_posted_date'], errors='coerce')
print(df_Germany['job_posted_date'].dtype)
```

```python
df_Germany['job_posted_month'] = df_Germany['job_posted_date'].dt.strftime('%B')
df_Germany.head()
```

```python

df_Germany_pivot = df_Germany.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_Germany_pivot
```

```python
# sort job_posted_month_name in df_Germany_pivot by job_posted_month value
df_Germany_pivot = df_Germany_pivot.reset_index()
df_Germany_pivot['job_posted_month_no'] = pd.to_datetime(df_Germany_pivot['job_posted_month'], format='%B').dt.month
df_Germany_pivot = df_Germany_pivot.sort_values('job_posted_month_no')
df_Germany_pivot = df_Germany_pivot.set_index('job_posted_month')
df_Germany_pivot = df_Germany_pivot.drop(columns='job_posted_month_no')

df_Germany_pivot
```

```python
df_Germany_pivot.plot(kind='line', figsize=(15, 10))
```

```python

Top_5_Jobs = df_Germany['job_title_short'].value_counts().head(5)
Top_5_Jobs = Top_5_Jobs.index.tolist()
```

```python

df_Germany_pivot[Top_5_Jobs].plot(kind='line', figsize=(15, 10))
plt.show()
plt.xlabel(2023)
plt.ylabel('Job_count')
plt.title('Top 5 Job Titles in Germany')
plt.legend()
```
