```python
import pandas as pd
import os
import zipfile

# Define the zip file path
zip_file_path = "/mnt/data/archive (13).zip"
extract_folder = "/mnt/data/extracted_files"

# Extract the ZIP file
with zipfile.ZipFile(r"C:\Users\mosta\Downloads\archive (13).zip", "r") as zip_ref:
    zip_ref.extractall(extract_folder)

# List all extracted files
csv_files = [f for f in os.listdir(extract_folder) if f.endswith(".csv")]

# Dictionary to store DataFrames
dfs = {}

# Read all CSV files into separate DataFrames
for file in csv_files:
    file_path = os.path.join(extract_folder, file)
    df_name = file.replace(".csv", "")  # Use file name as key
    dfs[df_name] = pd.read_csv(file_path)  # Read CSV file

# Print loaded DataFrames
print(f"Loaded {len(dfs)} CSV files: {list(dfs.keys())}")

# Example: Display first 5 rows of a specific DataFrame
print(dfs["blinkit_delivery_performance"].head())  # Change based on available file names
print(type(dfs))
```

```python
df_orders = dfs['blinkit_products'].merge(dfs['blinkit_order_items'], on='product_id', how='inner')
df_orders
```

```python

total_sold_units = df_orders.groupby('product_id')['quantity'].sum().reset_index()
total_sold_units.columns = ['product_id', 'total_sold_units']
df_orders = df_orders.merge(total_sold_units, on='product_id', how='left')
df_orders.head()
```

```python
df_orders = dfs['blinkit_products'].merge(dfs['blinkit_order_items'], on='product_id', how='inner')
df_orders
```

```python
df_orders['total_sales'] = df_orders['quantity'] * df_orders['price']
columns_to_drop = ['product_id', 'shelf_life_days', 'min_stock_level', 'margin_percentage', 'max_stock_level', 'mrp']
df_orders.drop(columns=[col for col in columns_to_drop if col in df_orders.columns], inplace=True)
df_orders
```

```python
pivot_table = df_orders.pivot_table(index=['category','product_name','brand'], values='total_sales', aggfunc='sum')
pivot_table
```

```python
pivot_sales= pivot_table['brand_total_sales'] = pivot_table.groupby('brand')['total_sales'].transform('sum')
pivot_sales
```

```python
pivot_category = pivot_table.pivot_table(index=['category'], values='total_sales', aggfunc='sum')
pivot_category
```

```python
pivot_brand = pivot_table.pivot_table(index='brand', columns='category', values='total_sales', aggfunc='sum')
Top_10_brands = pivot_brand.assign(Total_Sales=pivot_brand.sum(axis=1))\
.sort_values(by='Total_Sales', ascending=False)\
.head(10)


Top_10_brands.fillna(0, inplace=True)
Top_10_brands
```

```python
brand_contributions = pivot_table['total_sales'] / pivot_category['total_sales']*100
brand_contributions
```

```python
Top_20_brands = pivot_table.assign(Total_Sales=pivot_table.sum(axis=1))\
	.sort_values(by='Total_Sales', ascending=False)\
	.head(20)
Top_20_brands.drop(columns=['Total_Sales'], inplace=True)
Top_20_brands.drop(columns=['brand_total_sales'], inplace=True)
Top_20_brands
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
Top_10_brands = Top_10_brands.sort_values(by='Total_Sales', ascending=True)
plt.figure(figsize=(20, 15))
sns.set_style("whitegrid")
sns.barplot(x=Top_10_brands['Total_Sales'], y=Top_10_brands.index, palette="Blues_r")
plt.title("Top 10 Brands by Total Sales", fontsize=18, fontweight="bold")
plt.xlabel("Total Sales", fontsize=14, fontweight="bold")
plt.ylabel("Brand", fontsize=14, fontweight="bold")
for index, value in enumerate(Top_10_brands['Total_Sales']):
    plt.text(value + 500, index, f"{value:,.0f}", fontsize=12, fontweight="bold", verticalalignment="center")

sns.despine(left=True, bottom=True)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 8))
sns.set_style("whitegrid")

palette = sns.color_palette("coolwarm", len(pivot_category))


ax = sns.barplot(x=pivot_category.index, y=pivot_category['total_sales'], palette=palette)


plt.title("Total Sales by Category", fontsize=18, fontweight="bold", color="darkblue")
plt.xlabel("Category", fontsize=14, fontweight="bold")
plt.ylabel("Total Sales", fontsize=14, fontweight="bold")


plt.xticks(rotation=45, fontsize=12, fontweight="bold")


for p in ax.patches:
    ax.annotate(f"{p.get_height():,.0f}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight="bold", color="black")


sns.despine(left=True, bottom=True)


plt.show()
```
