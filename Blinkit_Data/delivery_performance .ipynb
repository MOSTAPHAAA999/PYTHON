```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert order time to datetime and extract hours
dfs['blinkit_orders']['order_time'] = pd.to_datetime(dfs['blinkit_orders']['order_date'])
dfs['blinkit_orders']['hour'] = dfs['blinkit_orders']['order_time'].dt.hour

# Filter delayed orders
delayed_orders = dfs['blinkit_delivery_performance'][dfs['blinkit_delivery_performance']['delivery_status'].str.contains('Delayed')]

# Merge to get hour column
delayed_orders = delayed_orders.merge(dfs['blinkit_orders'][['order_id', 'hour']], on='order_id', how='left')

# Count delayed orders per hour
time_slot_delays = delayed_orders['hour'].value_counts().sort_index()

# Identify peak delay hour
peak_delay_hour = time_slot_delays.idxmax()
max_delays = time_slot_delays.max()

print(f"Peak delay hour: {peak_delay_hour}:00 with {max_delays} delays.")

# Convert 24-hour format to 12-hour AM/PM format
def format_hour(hour):
    if hour == 0:
        return "12 AM"
    elif hour == 12:
        return "12 PM"
    elif hour < 12:
        return f"{hour} AM"
    else:
        return f"{hour - 12} PM"

# Create formatted x-axis labels
formatted_hours = [format_hour(h) for h in time_slot_delays.index]

# Plot setup
plt.figure(figsize=(20, 15))

# Define colors and highlight peak
colors = ["lightcoral" if h == peak_delay_hour else "steelblue" for h in time_slot_delays.index]

bars = plt.bar(time_slot_delays.index, time_slot_delays.values + 5, color=colors)  # Offset by +5 to move bars up

# Annotate peak delay hour
plt.text(peak_delay_hour, max_delays + 10, f"Peak: {max_delays}", 
         fontsize=12, fontweight="bold", ha="center", color="black")

# Labels and title
plt.xlabel("Hour of the Day", fontsize=14, fontweight="bold")
plt.ylabel("Number of Delayed Orders", fontsize=14, fontweight="bold")
plt.title("Delayed Orders per Time Slot", fontsize=16, fontweight="bold", color="darkblue")

# Format x-axis with 12-hour format
plt.xticks(ticks=time_slot_delays.index, labels=formatted_hours, rotation=45)

# Grid and layout adjustments
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.ylim(0, max_delays + 20)  # Adjust limit to prevent overlap

# Show final plot
plt.show()
```

```python
import pandas as pd

# Ensure 'actual_time' is in datetime format
dfs_delayed['actual_time'] = pd.to_datetime(dfs_delayed['actual_time'])

# Create a 'Year-Month' column
dfs_delayed['year_month'] = dfs_delayed['actual_time'].dt.to_period('M')

# Create a pivot table
pivot_table_date = dfs_delayed.pivot_table(index='year_month', 
                                      columns='delivery_status', 
                                      values='order_id', 
                                      aggfunc='count', 
                                      fill_value=0)

pivot_table_date
```

```python
import matplotlib.pyplot as plt

# Sort the pivot table by index in descending order
pivot_table_date = pivot_table_date.sort_index(ascending=False)

# Plot the sorted pivot table
pivot_table_date.plot(kind='barh', figsize=(20,15), title='Delayed Orders monthly')
plt.legend(title='Delivery Status', title_fontsize='13', fontsize='12', loc='upper right')
plt.show()
```
