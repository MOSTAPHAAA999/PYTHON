```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\mosta\Downloads\archive (16).zip"

df = pd.read_csv(file_path, compression='zip')


df
```

```python
df['Customer Remarks'].isnull()
df['Customer Remarks'].dropna(inplace=True)
df['Customer Remarks'].fillna('no answer', inplace=True)
```

```python
df['CSAT Score'].isnull()
df.isnull()
```

```python
df['order_date_time'] = pd.to_datetime(df['order_date_time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
df['order_date_time'].fillna('-',inplace=True)
df['order_date_time'].isnull().sum()
```

```python
df['Order_id'].dropna(inplace=True)
df['Order_id'].fillna('Not Available', inplace=True)
df['Order_id'].info()
```

```python
df['Customer_City'].isnull().sum()
df['Customer_City'].dropna(inplace=True)
df['Customer_City'].fillna('Not Available', inplace=True)   
df['Customer_City'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['Product_category'].dropna(inplace=True)
df['Product_category'].fillna('Not Available', inplace=True)
df['Product_category'].isnull().sum()
```

```python
avg_AHT = df['connected_handling_time' ].mean()
avg_AHT
avg_item_price = df['Item_price'].mean()
avg_item_price
```

```python
df['Item_price'].dropna(inplace=True)
df['Item_price'].fillna(avg_item_price, inplace=True)
df['Item_price'].isnull().sum()
```

```python
df['connected_handling_time'].dropna(inplace=True)
df['connected_handling_time'].fillna(avg_AHT, inplace=True)
df['connected_handling_time'].isnull().sum()
```
